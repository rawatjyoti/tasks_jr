# COVID2019 Analysis

Coronavirus disease is an infectious disease caused by newly discovered coronavirus.The virus is transmitted through droplets,sneezes,cough or exhales.These droplets are heavy to hung in the air, and quickly falls down on floors or surfaces.If anyone is in close proximity of someone who is suffering from COVID-19, there are chances to get infected. This disesase is first originated from Wuhan City,China and spread across the whole world. And also termed as pandemic by World Health Organisation(WHO).

# Maximum number of cases per day till date
Confirmed,recovered and deaths are grouped by date to get a day for the corresponding maximum cases.

```python
# Number of cases per day till date
casesperDayTillDate = df.groupby(["Date"])["Confirmed","Recovered","Deaths"]
casesperDayTillDate.max()
```

# Description of maximum number of cases 

```python
casesperDayTillDate = df.groupby(["Date"])["Confirmed","Recovered","Deaths"].max()
casesperDayTillDate.describe()
```

Maximum confirmed cases presented in dataset is 243453.
Maximum deaths cases presented here is 13915 which is really small if compared to number of confirmed cases.It depicts that death chances are minimum.
Maximum recovered from COVID19 are 63471.

```python
sb.countplot(x="Confirmed",hue="Deaths", data=casesperDayTillDate)
```

# Plotting cases per Day Till Date
plot type = bar
width= 50
height=10

```python

casesperDayTillDate.plot(kind="bar",figsize=(50,10))
```

# Plotting Deaths vs Confirmed

```python
plt.xlabel=("Deaths")
plt.ylabel=("Confirmed")
plt.scatter(casesperDayTillDate.Deaths,casesperDayTillDate.Confirmed, color="green", marker=".")
```

Observation :
Confirmed cases are showing progression with respect to number of deaths identified and attaining its maximum value till date.

# Plotting Recovered vs Confirmed

```python
plt.xlabel=("Recovered")
plt.ylabel=("Confirmed")
plt.scatter(casesperDayTillDate.Recovered,casesperDayTillDate.Confirmed, color="red", marker="+")
```

Here,confirmed cases shows gradual increase with respect to number of recovered cases .However it shoots up to the maximum when recovered cases approaches to 60000 .It also depicts a steady growth in confirmed cases from 10000 to 60000 recovered  cases.
